```python
There are three errors defined in arithmetic class 
OverflowError, ZeroDivisionError, FloatingPointError

try:
  arithmetic = 5/0
  print(arithmetic)
except ArithmeticError:
  print('You have just made an Arithmetic error')

try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
Lookup error class is used because LookupError Exception is the Base class for errors raised when something can't be found. 

pylenin_info = {'name': 'Lenin Mishra',
                'age': 28,
                'language': 'Python'}
user_input = input('What do you want to learn about Pylenin==> ')

try:
    print(f'{user_input} is {pylenin_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```
